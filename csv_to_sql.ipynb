# Setting up needed libraries 


```python
import pandas as pd
import sqlite3
import glob
```

# Printing our Csvs first lines in order to compare

```python
path = 'FitBit/*.csv'
csv_files = glob.glob(path)

for file in csv_files:
  
    df = pd.read_csv(file)
 
    print(f"File first line {file}:")
    print(df.iloc[0])
    print("\n")
```

# Connecting with SQLite 

```python
conn = sqlite3.connect('database.db')
cursor = conn.cursor()
```

# Converting CSV files into SQL tables 

```python
table_names = []

for file in csv_files: 
    df = pd.read_csv(file)
    table_name = file.split('\\')[-1].replace('.csv', '').replace('-', '_')
    df.to_sql(table_name, conn, if_exists='replace', index=False)
    table_names.append(table_name)

print("Table names on database:")
for name in table_names:
    print(name)
```

# Closing Database

```python
conn.close()
```
